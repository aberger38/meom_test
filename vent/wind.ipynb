```python
%run '/mnt/berger2/equipes/IGE/meom/workdir/berger2/fonctions/meom_fonctions/fonctions.ipynb'

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.netcdf as netcdf
import qgutils as qg
from matplotlib import ticker, cm
plt.ion()
```

```python
#paramètres gravés dans le marbre pour le stage

Delta = 10000                               #grid spacing
Delta_at = 16000
f0 = 9.37456E-05                           #paramètre de Coriolis
dh = np.array([350,750,2900])              #hauteurs des couches océaniques
dh_b = np.array([(dh[0]+dh[1])/2,(dh[0]+dh[1])/2])
g_prime = np.array([2.5E-02,1.25E-02])     #gravités réduites aux interfaces
Beta = 1.75360E-11
delta_ek = 5                               #ocean bottom Ekman thickness 
nu4 = 2E09                                 #bi-harmonic viscosity
Ht = np.sum(dh)
Ht_b = np.sum(dh_b)
toc = np.array([1.31693E+01,1.81693E+01,2.41693E+01])
bf = delta_ek*f0/(2*dh[-1])

#fréquences Brunt-vaisala

N_12 = g_prime[0]/((dh[0]+dh[1])/2)
N_23 = g_prime[1]/((dh[1]+dh[2])/2)
N2 = np.array([N_12,N_23])
```

```python
dir_abs  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2/qgcm-data/double_gyre_coupled/outdata_7_dt10/'
dir_rel  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2/qgcm-data/double_gyre_coupled/outdata_6tdiff_dt10/'


file0 = 'ocpo.nc'
file1 = 'ocsst.nc'
file_avge = 'avges.nc'
file_at = 'atpa.nc'
```

```python
#get fields psi
pfile_abs = dir_abs + file0
pfile_rel = dir_rel + file0

pfile_at_abs = dir_abs + file_at
pfile_at_rel = dir_rel + file_at

avge_file_abs = dir_abs + file_avge
avge_file_rel = dir_rel + file_avge


f_abs = netcdf.netcdf_file(dir_abs + file0,'r')
f_rel = netcdf.netcdf_file(dir_rel + file1,'r')
f_avge_abs = netcdf.netcdf_file(avge_file_abs,'r')
f_avge_rel = netcdf.netcdf_file(avge_file_rel,'r')
```

```python
N2,f0 = qg.reshape3d(dh,N2,f0)

si_t = f_abs.variables['time'][:].copy().size

p_me_at = qg.interp_on_c(f_avge_abs.variables['pa'][:,:].copy())/f0
nl,N,naux = p_me_at.shape

test = si_t

average = True
```

```python
p_me_at_abs = qg.interp_on_c(f_avge_abs.variables['pa'][:,:].copy())/f0
p_me_at_rel = qg.interp_on_c(f_avge_rel.variables['pa'][:,:].copy())/f0


u_me_at_abs,v_me_at_abs = qg.comp_vel(p_me_at_abs, Delta_at, loc='center')
u_me_at_rel,v_me_at_rel = qg.comp_vel(p_me_at_rel, Delta_at, loc='center')
```

```python
%%time

#physical map

var_at_abs = np.zeros((nl,N,naux))
var_at_rel = np.zeros((nl,N,naux))

n_me=1

for it in range(0,test):
    
    p_abs = qg.read_qgcm(pfile_at_abs, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    p_rel = qg.read_qgcm(pfile_at_abs, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True) 
    
    u_abs,v_abs = qg.comp_vel(p_abs, Delta=Delta, loc='center')
    u_rel,v_rel = qg.comp_vel(p_rel, Delta=Delta, loc='center')
  
    
    u_p_at_abs = u_abs-u_me_at_abs
#    v_p_at_abs = v_abs-v_me_at_abs
    u_p_at_rel = u_rel-u_me_at_rel
#    v_p_at_rel = v_rel-v_me_at_rel

    var_at_abs += (u_p_at_abs**2 - var_at_abs)/n_me
    var_at_rel += (u_p_at_rel**2 - var_at_rel)/n_me
    
    n_me += 1
    
```
