```python
%run '/mnt/berger2/equipes/IGE/meom/workdir/berger2/fonctions/meom_fonctions/fonctions.ipynb'

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.netcdf as netcdf
import qgutils as qg
from matplotlib import ticker, cm
plt.ion()
```

```python
#paramètres gravés dans le marbre pour le stage

Delta = 10000                               #grid spacing
f0 = 9.37456E-05                           #paramètre de Coriolis
dh = np.array([350,750,2900])              #hauteurs des couches océaniques
dh_b = np.array([(dh[0]+dh[1])/2,(dh[0]+dh[1])/2])
g_prime = np.array([2.5E-02,1.25E-02])     #gravités réduites aux interfaces
Beta = 1.75360E-11
delta_ek = 5                               #ocean bottom Ekman thickness 
nu4 = 2E09                                 #bi-harmonic viscosity
Ht = np.sum(dh)
Ht_b = np.sum(dh_b)
toc = np.array([1.31693E+01,1.81693E+01,2.41693E+01])
bf = delta_ek*f0/(2*dh[-1])

#fréquences Brunt-vaisala

N_12 = g_prime[0]/((dh[0]+dh[1])/2)
N_23 = g_prime[1]/((dh[1]+dh[2])/2)
N2 = np.array([N_12,N_23])
```

```python
dir0  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2/qgcm-data/double_gyre_coupled/outdata_6/'
dir1  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2/qgcm-data/double_gyre_coupled/outdata_9tdiff/'

file0 = 'ocpo.nc'
file1 = 'ocsst.nc'
```

```python
#get fields psi
pfile0 = dir0 + file0
pfile1 = dir1 + file0

#wind forcing
forcing_z = dir0 + file1

#buyoancy forcing
forcing_b = dir0 + file1

f1 = netcdf.netcdf_file(dir0 + file0,'r')
f2 = netcdf.netcdf_file(dir0 + file1,'r')
```

```python
N2,f0 = qg.reshape3d(dh,N2,f0)

si_t = f1.variables['time'][:].copy().size

p = qg.read_qgcm(pfile0, 0, var='p', rescale = 1/f0, interp = True, subtract_bc = False )
nl,N,naux = p.shape

test = si_t

average = True
```

```python
# flot moyen abs
flot_me = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_abs/termes_flot_moyen_abs.npz')
u_me = flot_me['u_me']
v_me = flot_me['v_me']

flot_me_bis = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_abs/etapes_flot_moyen_abs.npz')
         
p_me = flot_me_bis['p_me']
b_me = flot_me_bis['b_me']
```

```python
#flot moyen rel
flot_me = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_rel/termes_flot_moyen_rel.npz')
u_me_rel = flot_me['u_me']
v_me_rel = flot_me['v_me']

flot_me_bis = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_rel/etapes_flot_moyen_rel.npz')
         
p_me_rel = flot_me_bis['p_me']
b_me_rel = flot_me_bis['b_me']
```

```python
%%time

# compute all terms of dissipation

e_dissip_p_me = np.zeros((nl,N,N))
e_bottom_p_me = np.zeros((N,N))
u_p_2 = np.zeros((nl,N,N))
v_p_2 = np.zeros((nl,N,N))

n_me=1

for it in range(0,test):
    
    p = qg.read_qgcm(pfile, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    
    u,v = qg.comp_vel(p, Delta=Delta, loc='center')
  
    
    u_p = u-u_me
    v_p = v-v_me
    
    # termes pour dissipation phys
    dissip_ke_phys = -nu4*(u_p*qg.laplacian(qg.laplacian(u_p,Delta),Delta)
                     + v_p*qg.laplacian(qg.laplacian(v_p,Delta),Delta))
                           
    # mean for saving (physical)
    e_dissip_p_me += (dissip_ke_phys - e_dissip_p_me)/n_me
    e_bottom_p_me += (u_p[2]**2+v_p[2]**2 - e_bottom_p_me)/n_me
    u_p_2 += (u_p**2 - u_p_2)/n_me
    v_p_2 += (v_p**2 - u_p_2)/n_me
    
    
    
    n_me += 1
    
```

```python
%%time

# compute rec pe abs

pe_me2pe_p_me = np.zeros((nl-1,N,N))

p_me_bis = np.zeros((nl-1,N,N))

p_me_bis[0] = (p_me[0]*dh[0]+ p_me[1]*dh[1])/(dh[0]+dh[1])
p_me_bis[1] = (p_me[2]*dh[2]+ p_me[1]*dh[1])/(dh[2]+dh[1])

n_me=1

for it in range(0,test):
    
    p = qg.read_qgcm(pfile0, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    
    b = qg.p2b(p, dh, f0)
    
    
    p_bis = np.zeros((nl-1,N,N))

    p_bis[0] = (p[0]*dh[0]+ p[1]*dh[1])/(dh[0]+dh[1])
    p_bis[1] = (p[2]*dh[2]+ p[1]*dh[1])/(dh[2]+dh[1])
    
    
    
    p_p = p_bis-p_me_bis
    b_p = b-b_me
    
    rec_pe = (1/N2)*b_me*qg.jacobian(p_p, b_p, Delta)
    pe_me2pe_p_me += (-rec_pe - pe_me2pe_p_me)/n_me
    
    
    n_me += 1
```

```python
%%time

# compute rec pe rel

pe_me2pe_p_me_rel = np.zeros((nl-1,N,N))

p_me_bis_rel = np.zeros((nl-1,N,N))

p_me_bis_rel[0] = (p_me_rel[0]*dh[0]+ p_me_rel[1]*dh[1])/(dh[0]+dh[1])
p_me_bis_rel[1] = (p_me_rel[2]*dh[2]+ p_me_rel[1]*dh[1])/(dh[2]+dh[1])

n_me=1

for it in range(0,test):
    
    p_rel = qg.read_qgcm(pfile1, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    
    b_rel = qg.p2b(p_rel, dh, f0)
    
    
    p_bis_rel = np.zeros((nl-1,N,N))

    p_bis_rel[0] = (p_rel[0]*dh[0]+ p_rel[1]*dh[1])/(dh[0]+dh[1])
    p_bis_rel[1] = (p_rel[2]*dh[2]+ p_rel[1]*dh[1])/(dh[2]+dh[1])
    
    
    
    p_p_rel = p_bis_rel-p_me_bis_rel
    b_p_rel = b_rel-b_me_rel
    
    rec_pe_rel = (1/N2)*b_me_rel*qg.jacobian(p_p_rel, b_p_rel, Delta)
    pe_me2pe_p_me_rel += (-rec_pe_rel - pe_me2pe_p_me_rel)/n_me
    
    
    n_me += 1
```

```python
#mean PE to turbulent PE


pe_me2pe_p_me_i = integral_z(pe_me2pe_p_me, dh_b, Delta, Ht_b)
maxi = max(np.max(pe_me2pe_p_me_i),np.abs(np.min(pe_me2pe_p_me_i)))


#maxi = 7.8e-7

fig, ax1 = plt.subplots(1, 1, figsize = (8,6),tight_layout=True)


col = ax1.imshow(pe_me2pe_p_me_i, extent= [0,5000,0,5000], cmap='RdBu_r', vmax = maxi/15, vmin = -maxi/15)
ax1.set_aspect('equal')
#ax1.set_title('abs')
ax1.set_xlabel('x (km)')
ax1.set_ylabel('y (km)')

m = plt.cm.ScalarMappable(cmap='RdBu_r')
m.set_array(pe_me2pe_p_me)
m.set_clim(-maxi/15, maxi/15)
plt.colorbar(m, ax= ax1).set_label(r'$m^2.s^{-3}$')

#savefig
plt.savefig('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/figs_abs/pe_me2pe_p_me.png')
```

```python
pe_me2pe_p_me_i_rel = integral_z(pe_me2pe_p_me_rel, dh_b, Delta, Ht_b)

fig, ax2 = plt.subplots(1, 1, figsize = (8,6),tight_layout=True)

col = ax2.imshow(pe_me2pe_p_me_i_rel, extent= [0,5000,0,5000], cmap='RdBu_r', vmax = maxi/15, vmin = -maxi/15)
ax2.set_aspect('equal')
#ax2.set_title('rel')
ax2.set_xlabel('x (km)')
ax2.set_ylabel('y (km)')

m = plt.cm.ScalarMappable(cmap='RdBu_r')
m.set_array(pe_me2pe_p_me_rel)
m.set_clim(-maxi/15, maxi/15)
plt.colorbar(m, ax= ax2).set_label(r'$m^2.s^{-3}$')

#savefig
plt.savefig('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/figs_rel/pe_me2pe_p_me.png')
```
