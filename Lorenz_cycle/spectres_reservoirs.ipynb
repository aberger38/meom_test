```python
%run '/mnt/berger2/equipes/IGE/meom/workdir/berger2/fonctions/meom_fonctions/fonctions.ipynb'

import numpy as np
import matplotlib.pyplot as plt
import scipy.io.netcdf as netcdf
import qgutils as qg
from matplotlib import ticker, cm
plt.ion()
```

```python
#paramètres gravés dans le marbre pour le stage

Delta = 10000                               #grid spacing
f0 = 9.37456E-05                           #paramètre de Coriolis
dh = np.array([350,750,2900])              #hauteurs des couches océaniques
dh_b = np.array([(dh[0]+dh[1])/2,(dh[0]+dh[1])/2])
g_prime = np.array([2.5E-02,1.25E-02])     #gravités réduites aux interfaces
Beta = 1.75360E-11
delta_ek = 5                               #ocean bottom Ekman thickness 
nu4 = 2E09                                 #bi-harmonic viscosity
Ht = np.sum(dh)
Ht_b = np.sum(dh_b)
toc = np.array([1.31693E+01,1.81693E+01,2.41693E+01])
bf = delta_ek*f0/(2*dh[-1])

#fréquences Brunt-vaisala

N_12 = g_prime[0]/((dh[0]+dh[1])/2)
N_23 = g_prime[1]/((dh[1]+dh[2])/2)
N2 = np.array([N_12,N_23])
```

```python
dir0  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2//qgcm-data/double_gyre_coupled/outdata_8newent/'
dir1  = '/mnt/berger2/equipes/IGE/meom/workdir/berger2//qgcm-data/double_gyre_coupled/outdata_7tdiff_newent/'
file0 = 'ocpo.nc'
```

```python
#get xp, yp : basin size
f1 = netcdf.netcdf_file(dir0 + file0,'r')

xp = f1.variables['xp'][:-1].copy()
yp = f1.variables['xp'][:-1].copy()
```

```python
#load absolute wind run

flot_me_et = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_abs/etapes_flot_moyen_abs.npz')
p_me_abs= flot_me_et['p_me']



#load relative wind run

flot_me_et_rel = np.load('/mnt/berger2/equipes/IGE/meom/workdir/berger2/git/meom_test/Lorenz_cycle/tableaux_termes_rel/etapes_flot_moyen_rel.npz')
p_me_rel= flot_me_et_rel['p_me']




```

```python
pfile_abs = dir0 + file0
pfile_rel = dir1 + file0
```

```python
def forward(x):
    return 1/x

def inverse(x):
    return 1/x
```

```python
nl,N,naux = p_me_abs.shape
```

```python
k,fl = qg.get_spec_flux(psi1 = p_me_abs,psi2 = p_me_abs,Delta=Delta)
#fl_z = integral_z(psi = fl, dh=dh, Delta=Delta, Ht=Ht)
z, n_k = fl.shape
```

```python
si_t = f1.variables['time'][:].copy().size
```

```python
%%time

fl_ke_p_abs_me = np.zeros((nl,n_k))
fl_ke_p_rel_me = np.zeros((nl,n_k))
fl_pe_p_abs_me = np.zeros((nl,n_k))
fl_pe_p_rel_me = np.zeros((nl,n_k))

n_me = 1

test = 5

for it in range(0,si_t):
    
    p_abs = qg.read_qgcm(pfile_abs, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    p_rel = qg.read_qgcm(pfile_rel, it, var='p', rescale = 1/f0, interp = True, subtract_bc = True)
    
    p_p_abs = p_abs - p_me_abs
    p_p_rel = p_rel - p_me_rel
    
    lap_p_abs = qg.laplacian(p_p_abs, Delta)
    lap_p_rel = qg.laplacian(p_p_rel, Delta)
    
    gamma_p_abs = qg.p2stretch(p_p_abs,dh,N2,f0)
    gamma_p_rel = qg.p2stretch(p_p_rel,dh,N2,f0)
    
    #flux ke et pe pour les 2 runs
    
    k_ke_p_abs, fl_ke_p_abs = qg.get_spec_1D(p_p_abs, lap_p_abs, Delta)
    fl_ke_p_abs_me += (fl_ke_p_abs-fl_ke_p_abs_me)/n_me
    
    k_ke_p_rel, fl_ke_p_rel = qg.get_spec_1D(p_p_rel, lap_p_rel, Delta)
    fl_ke_p_rel_me += (fl_ke_p_rel-fl_ke_p_rel_me)/n_me
    
    k_pe_p_abs, fl_pe_p_abs = qg.get_spec_1D(p_p_abs, gamma_p_abs, Delta)
    fl_pe_p_abs_me += (fl_pe_p_abs-fl_pe_p_abs_me)/n_me
    
    k_pe_p_rel, fl_pe_p_rel = qg.get_spec_1D(p_p_rel, gamma_p_rel, Delta)
    fl_pe_p_rel_me += (fl_pe_p_rel-fl_pe_p_rel_me)/n_me
    
    n_me += 1
```

```python
# flux resrvoirs d'énergie du flot moyen

lap_abs = qg.laplacian(p_abs, Delta)
lap_rel = qg.laplacian(p_rel, Delta)
    
gamma_abs = qg.p2stretch(p_abs,dh,N2,f0)
gamma_rel = qg.p2stretch(p_rel,dh,N2,f0)
    
#flux ke et pe pour les 2 runs
    
k_ke_abs, fl_ke_abs = qg.get_spec_1D(p_abs, lap_abs, Delta)
    
k_ke_rel, fl_ke_rel = qg.get_spec_1D(p_rel, lap_rel, Delta)
    
k_pe_abs, fl_pe_abs = qg.get_spec_1D(p_abs, gamma_abs, Delta)
    
k_pe_rel, fl_pe_rel = qg.get_spec_1D(p_rel, gamma_rel, Delta)
    
    
```

```python
fl_ke_abs_i = integral_z(psi = fl_ke_abs, dh = dh, Delta = Delta, Ht = Ht)

fl_pe_abs_i = integral_z(psi = fl_pe_abs, dh = dh, Delta = Delta, Ht = Ht)

fl_ke_rel_i = integral_z(psi = fl_ke_rel, dh = dh, Delta = Delta, Ht = Ht)

fl_pe_rel_i = integral_z(psi = fl_pe_rel, dh = dh, Delta = Delta, Ht = Ht)

fl_ke_p_abs_i = integral_z(psi = fl_ke_p_abs_me, dh = dh, Delta = Delta, Ht = Ht)

fl_pe_p_abs_i = integral_z(psi = fl_pe_p_abs_me, dh = dh, Delta = Delta, Ht = Ht)

fl_ke_p_rel_i = integral_z(psi = fl_ke_p_rel_me, dh = dh, Delta = Delta, Ht = Ht)

fl_pe_p_rel_i = integral_z(psi = fl_pe_p_rel_me, dh = dh, Delta = Delta, Ht = Ht)
```
